```python
import torch
import torch.nn as nn
import torchvision.datasets as datasets
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
```

```python
# Q5.1
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (1.0,))
])

train_dataset = datasets.MNIST(root='./data', train=True, download=True, transform=transform)
test_dataset = datasets.MNIST(root='./data', train=False, download=True, transform=transform)

train_loader = torch.utils.data.DataLoader(dataset=train_dataset, batch_size=200, shuffle=True)
test_loader = torch.utils.data.DataLoader(dataset=test_dataset, batch_size=200, shuffle=False)

class MLP(nn.Module):
    def __init__(self, depth, total_hidden_neurons):
        super(MLP, self).__init__()
        self.depth = depth
        if self.depth != 0:
            self.hidden_size = total_hidden_neurons // self.depth
        if self.depth == 0:
            self.model = nn.Sequential(nn.Linear(28 * 28, 10))
        else:
            layers = []
            layers.append(nn.Linear(28 * 28, self.hidden_size))
            layers.append(nn.ReLU())
            if self.depth > 1:
                for _ in range(self.depth - 1):
                    layers.append(nn.Linear(self.hidden_size, self.hidden_size))
                    layers.append(nn.ReLU())
            layers.append(nn.Linear(self.hidden_size, 10))
            self.model = nn.Sequential(*layers)

    def forward(self, x):
        x = x.view(-1, 28 * 28)
        return self.model(x)

depths = [0, 1, 2, 3, 4]
total_hidden_neurons = 100
learning_rate = 0.1

for depth in depths:
    print(f"Depth: {depth}")

    model = MLP(depth, total_hidden_neurons)
    criterion = nn.CrossEntropyLoss()
    optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)

    # Training loop
    num_epochs = 10

    train_losses = []
    test_losses = []
    train_accuracies = []
    test_accuracies = []

    for epoch in range(num_epochs):
        model.train()
        correct_train = 0
        total_train = 0
        running_train_loss = 0.0

        for i, (images, labels) in enumerate(train_loader):
            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_train_loss += loss.item()
            _, predicted = torch.max(outputs.data, 1)
            total_train += labels.size(0)
            correct_train += (predicted == labels).sum().item()

        train_accuracy = 100 * correct_train / total_train
        train_accuracies.append(train_accuracy)
        train_loss = running_train_loss / len(train_loader)
        train_losses.append(train_loss)

        model.eval()
        correct_test = 0
        total_test = 0
        running_test_loss = 0.0

        with torch.no_grad():
            for images, labels in test_loader:
                outputs = model(images)
                loss = criterion(outputs, labels)
                running_test_loss += loss.item()
                _, predicted = torch.max(outputs.data, 1)
                total_test += labels.size(0)
                correct_test += (predicted == labels).sum().item()

        test_accuracy = 100 * correct_test / total_test
        test_accuracies.append(test_accuracy)
        test_loss = running_test_loss / len(test_loader)
        test_losses.append(test_loss)

        print(f"Epoch [{epoch+1}/{num_epochs}], "
              f"Train Loss: {train_loss:.4f}, Train Accuracy: {train_accuracy:.2f}%, "
              f"Test Loss: {test_loss:.4f}, Test Accuracy: {test_accuracy:.2f}%")

    # torch.save(model.state_dict(), f"model_depth_{depth}.pth")

    plt.figure(figsize=(10, 5))
    # Plotting the training and test loss
    plt.subplot(1, 2, 1)
    plt.plot(range(1, num_epochs + 1), train_losses, label='Train Loss')
    plt.plot(range(1, num_epochs + 1), test_losses, label='Test Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title('Training and Test Loss')
    plt.legend()
    plt.show()

    # Plotting the training and test accuracy
    plt.figure(figsize=(10, 5))
    plt.plot(range(1, num_epochs + 1), train_accuracies, label='Train Accuracy')
    plt.plot(range(1, num_epochs + 1), test_accuracies, label='Test Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.title('Training and Test Accuracy')
    plt.legend()
    plt.show()
```

```python
# Q5.2
class MLP(nn.Module):
    def __init__(self, depth, total_hidden_neurons):
        super(MLP, self).__init__()
        self.depth = depth
        if self.depth != 0:
            self.hidden_size = total_hidden_neurons // self.depth
        if self.depth == 0:
            self.hidden_size = 0
            self.model = nn.Sequential(nn.Linear(28 * 28, 10))
        else:
            layers = []
            layers.append(nn.Linear(28 * 28, self.hidden_size))
            layers.append(nn.ReLU())
            if self.depth > 1:
                for _ in range(self.depth - 1):
                    layers.append(nn.Linear(self.hidden_size, self.hidden_size))
                    layers.append(nn.ReLU())
            layers.append(nn.Linear(self.hidden_size, 10))
            self.model = nn.Sequential(*layers)

    def forward(self, x):
        x = x.view(-1, 28 * 28)
        return self.model(x)

    def get_hidden_size(self):
        return self.hidden_size

def count_parameters(depth, hidden_size):
    input_size = 28 * 28
    output_size = 10

    if depth == 0:
        total_params = input_size * output_size
    else:
        total_params = input_size * hidden_size + hidden_size * hidden_size * (depth - 1) + hidden_size * output_size

    return total_params

def count_bias(depth, hidden_size):
    input_size = 28 * 28
    output_size = 10

    if depth == 0:
        total_params = input_size * output_size + output_size
    else:
        total_params = input_size * hidden_size + hidden_size * hidden_size * (depth - 1) + hidden_size * output_size + hidden_size * depth + output_size

    return total_params


depths = [0, 1, 2, 3, 4]
total_hidden_neurons = 100

for depth in depths:
    # load model
    model = MLP(depth, total_hidden_neurons)
    hidden_size = model.get_hidden_size()

    # count params
    total_params = count_parameters(depth, hidden_size)
    total_params_bias = count_bias(depth, hidden_size)
    print(f"Depth {depth} - Total Parameters: {total_params} - Total Parameters with Bias: {total_params_bias}")
```

```python
# Q5.3
def calculate_hidden_size(depth, total_parameters):
    input_size = 28 * 28
    output_size = 10

    a = depth - 1
    b = input_size + output_size
    c = -total_parameters

    discriminant = b**2 - 4*a*c

    # Use the quadratic formula to solve for hidden_size
    if depth == 1:
        return int(total_parameters / (input_size + output_size))
    else:
        sqrt_discriminant = discriminant ** 0.5
        hidden_size_1 = (-b + sqrt_discriminant) / (2*a)
        hidden_size_2 = (-b - sqrt_discriminant) / (2*a)
        return int(hidden_size_1) if hidden_size_1 > 0 else int(hidden_size_2)

transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (1.0,))
])

train_dataset = datasets.MNIST(root='./data', train=True, download=True, transform=transform)
test_dataset = datasets.MNIST(root='./data', train=False, download=True, transform=transform)

train_loader = torch.utils.data.DataLoader(dataset=train_dataset, batch_size=200, shuffle=True)
test_loader = torch.utils.data.DataLoader(dataset=test_dataset, batch_size=200, shuffle=False)

class MLP(nn.Module):
    def __init__(self, depth, hidden_size):
        super(MLP, self).__init__()
        self.depth = depth
        self.hidden_size = hidden_size
        layers = []
        layers.append(nn.Linear(28 * 28, self.hidden_size))
        layers.append(nn.ReLU())
        if self.depth > 1:
            for _ in range(self.depth - 1):
                layers.append(nn.Linear(self.hidden_size, self.hidden_size))
                layers.append(nn.ReLU())
        layers.append(nn.Linear(self.hidden_size, 10))
        self.model = nn.Sequential(*layers)

    def forward(self, x):
        x = x.view(-1, 28 * 28)
        return self.model(x)

depths = [1, 2, 3, 4]
learning_rate = 0.1

for depth in depths:
    # Total parameters for a single hidden layer MLP with 100 hidden neurons
    single_layer_total_params = 79400

    # Calculate neurons for each hidden layer to maintain the same total parameters
    neurons_per_layer = calculate_hidden_size(depth, single_layer_total_params)

    print(f"Depth: {depth}, Neurons per Hidden Layer: {neurons_per_layer}")

    # Proceed with training the MLP with the determined number of neurons per hidden layer
    model = MLP(depth, neurons_per_layer)
    criterion = nn.CrossEntropyLoss()
    optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)

    # Training loop
    num_epochs = 10

    train_losses = []
    test_losses = []
    train_accuracies = []
    test_accuracies = []

    for epoch in range(num_epochs):
        model.train()
        correct_train = 0
        total_train = 0
        running_train_loss = 0.0

        for i, (images, labels) in enumerate(train_loader):
            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_train_loss += loss.item()
            _, predicted = torch.max(outputs.data, 1)
            total_train += labels.size(0)
            correct_train += (predicted == labels).sum().item()

        train_accuracy = 100 * correct_train / total_train
        train_accuracies.append(train_accuracy)
        train_loss = running_train_loss / len(train_loader)
        train_losses.append(train_loss)

        model.eval()
        correct_test = 0
        total_test = 0
        running_test_loss = 0.0

        with torch.no_grad():
            for images, labels in test_loader:
                outputs = model(images)
                loss = criterion(outputs, labels)
                running_test_loss += loss.item()
                _, predicted = torch.max(outputs.data, 1)
                total_test += labels.size(0)
                correct_test += (predicted == labels).sum().item()

        test_accuracy = 100 * correct_test / total_test
        test_accuracies.append(test_accuracy)
        test_loss = running_test_loss / len(test_loader)
        test_losses.append(test_loss)

        print(f"Epoch [{epoch+1}/{num_epochs}], "
              f"Train Loss: {train_loss:.4f}, Train Accuracy: {train_accuracy:.2f}%, "
              f"Test Loss: {test_loss:.4f}, Test Accuracy: {test_accuracy:.2f}%")

    plt.figure(figsize=(10, 5))
    # Plotting the training and test loss
    plt.subplot(1, 2, 1)
    plt.plot(range(1, num_epochs + 1), train_losses, label='Train Loss')
    plt.plot(range(1, num_epochs + 1), test_losses, label='Test Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title('Training and Test Loss')
    plt.legend()
    plt.show()

    # Plotting the training and test accuracy
    plt.figure(figsize=(10, 5))
    plt.plot(range(1, num_epochs + 1), train_accuracies, label='Train Accuracy')
    plt.plot(range(1, num_epochs + 1), test_accuracies, label='Test Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.title('Training and Test Accuracy')
    plt.legend()
    plt.show()
```

```python
# Q5.4
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (1.0,))
])

train_dataset = datasets.MNIST(root='./data', train=True, download=True, transform=transform)
test_dataset = datasets.MNIST(root='./data', train=False, download=True, transform=transform)

train_loader = torch.utils.data.DataLoader(dataset=train_dataset, batch_size=200, shuffle=True)
test_loader = torch.utils.data.DataLoader(dataset=test_dataset, batch_size=200, shuffle=False)

class MLP(nn.Module):
    def __init__(self, config):
        super(MLP, self).__init__()
        self.config = config
        layers = []
        layers.append(nn.Linear(28 * 28, self.config[0]))
        layers.append(nn.ReLU())
        layers.append(nn.Linear(self.config[0], self.config[1]))
        layers.append(nn.ReLU())
        layers.append(nn.Linear(self.config[1], self.config[2]))
        layers.append(nn.ReLU())
        layers.append(nn.Linear(self.config[2], self.config[3]))
        layers.append(nn.ReLU())
        layers.append(nn.Linear(self.config[3], 10))
        self.model = nn.Sequential(*layers)

    def forward(self, x):
        x = x.view(-1, 28 * 28)
        return self.model(x)

hidden_neurons_combinations = [
    [25, 25, 25, 25],
    [10, 30, 40, 20],
    [40, 30, 20, 10],
    [40, 30, 15, 15],
    [40, 20, 20, 20],
    [15, 15, 30, 40],
    [20, 30, 40, 10],
    [5, 20, 50, 25],
    [35, 40, 20, 5]
]

learning_rate = 0.1

best_config = None
best_train_accuracy = 0.0
best_test_accuracy = 0.0

for config in hidden_neurons_combinations:

    # print(f"Configration: {config}")

    model = MLP(config)
    criterion = nn.CrossEntropyLoss()
    optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)

    # Training loop
    num_epochs = 10

    train_losses = []
    test_losses = []
    train_accuracies = []
    test_accuracies = []

    for epoch in range(num_epochs):
        model.train()
        correct_train = 0
        total_train = 0
        running_train_loss = 0.0

        for i, (images, labels) in enumerate(train_loader):
            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_train_loss += loss.item()
            _, predicted = torch.max(outputs.data, 1)
            total_train += labels.size(0)
            correct_train += (predicted == labels).sum().item()

        train_accuracy = 100 * correct_train / total_train
        train_accuracies.append(train_accuracy)
        train_loss = running_train_loss / len(train_loader)
        train_losses.append(train_loss)

        model.eval()
        correct_test = 0
        total_test = 0
        running_test_loss = 0.0

        with torch.no_grad():
            for images, labels in test_loader:
                outputs = model(images)
                loss = criterion(outputs, labels)
                running_test_loss += loss.item()
                _, predicted = torch.max(outputs.data, 1)
                total_test += labels.size(0)
                correct_test += (predicted == labels).sum().item()

        test_accuracy = 100 * correct_test / total_test
        test_accuracies.append(test_accuracy)
        test_loss = running_test_loss / len(test_loader)
        test_losses.append(test_loss)

    # print('Train Accuracy:', train_accuracies[-1])
    # print('Test Accuracy:', test_accuracies[-1])
    # print(f"Configration {config} complete.")

    if test_accuracies[-1] > best_test_accuracy:
        best_config = config
        best_train_accuracy = train_accuracies[-1]
        best_test_accuracy = test_accuracies[-1]

print(f'Best Configration: {best_config}')
print(f'Train Accuracy: {best_train_accuracy:.2f}')
print(f'Test Accuracy: {best_test_accuracy:.2f}')
```
